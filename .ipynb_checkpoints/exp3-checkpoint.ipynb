```python
import nltk
from nltk.corpus import stopwords, gutenberg
from nltk.tokenize import word_tokenize, sent_tokenize
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import re
import string
```

```python
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('stopwords')
corpus = gutenberg.raw('austen-emma.txt')
```

```python
stop_words = set(stopwords.words('english'))
sentences = sent_tokenize(corpus)
processed_data = []

for sent in sentences:
    sent = re.sub(r'[^a-zA-Z ]', '', sent)
    tokens = word_tokenize(sent.lower())
    filtered = [w for w in tokens if w not in stop_words and w.isalpha()]
    processed_data.append(filtered)
```

```python
model = Word2Vec(sentences=processed_data,
                 vector_size=100,
                 window=5,
                 min_count=2,
                 workers=4,
                 sg=1)  # sg=0 for CBOW, sg=1 for Skip-gram
```

```python
model.wv.most_similar('emma')
```

```python
model.wv.similarity('emma', 'harriet')
```

```python
import numpy as np

words = list(model.wv.index_to_key)[:100]
word_vectors = [model.wv[w] for w in words]

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(np.array(word_vectors))

plt.figure(figsize=(12, 8))
for i, word in enumerate(words):
    plt.scatter(X_tsne[i, 0], X_tsne[i, 1])
    plt.annotate(word, (X_tsne[i, 0], X_tsne[i, 1]))
plt.title("t-SNE Visualization of Word2Vec Embeddings")
plt.show()
```
